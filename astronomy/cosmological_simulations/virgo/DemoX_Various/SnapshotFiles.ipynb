# Millennium snapshot files
This notebook has some simple functions to identify the locations of Millennium snapshot files over the complex virgo_filedb volume.

```python
import glob
import re
import pandas
```

```python
def fileRank(file):
    return int(file.split(".")[-1])

def snapnum(file):
    m=re.search("_(\d\d\d).",file)
    return int(m.groups()[-1])
#    return int(m.group())

def filedbloc(file):
    file= file.replace('/home/idies/workspace/virgo_filedb/data','')
    digits=re.findall(r'\d+',file)
    return "//filedb"+digits[0]+'/data'+digits[1]+file[5:]

def volume(file):
    m=re.search('//filedb(\d\d)/data(\d\d)',file)
    return int(m.group(1))*100+int(m.group(2))
```

```python
def snapshotsorter(f):
    return fileRank(f)

def snapdirsorter(f):
    return int(snapnum(f)*1000)+int(fileRank(f))
```

```python
def snapshot(snapnum,sim='Millennium'):
    snap='%03d'%snapnum
    files=glob.glob('/home/idies/workspace/virgo_filedb/*/{sim}/snapdir_{snap}/snap*.*'.format(snap=snap,sim=sim)
                    , recursive=True)
    files.sort(key=snapshotsorter)
    return files

def subhaloids(snapnum,sim='Millennium'):
    snap='%03d'%snapnum
    idsfiles=glob.glob('/home/idies/workspace/virgo_filedb/*/{sim}/postproc_{snap}/sub_ids_*.*'.format(snap=snap,sim=sim)
                    , recursive=True)
    idsfiles.sort(key=snapshotsorter)
    return idsfiles

def allsnapshots(sim="Millennium"):
    files=glob.glob('/home/idies/workspace/virgo_filedb/*/{sim}/snapdir_*/snap*.*'.format(sim=sim)
                    , recursive=True)
    files.sort(key=snapdirsorter)
    return files

def allsubhaloids(sim="Millennium"):
    files=glob.glob('/home/idies/workspace/virgo_filedb/*/{sim}/postproc_*/sub_ids_*.*'.format(sim=sim)
                    , recursive=True)
    files.sort(key=snapdirsorter)
    return files
```

```python
allfiles=snapshot(13,'Millennium')
subhaloidsfiles=subhaloids(63,'Millennium')
```

```python
snaps=[snapnum(f) for f in allfiles]
ranks=[fileRank(f) for f in allfiles]
locations=[filedbloc(f) for f in allfiles]
volumes=[volume(f) for f in locations]
```
